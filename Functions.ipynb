1. What is the difference between a function and a method in Python?

=>  both perform specific tasks, but they differ in their context and usage. Functions are standalone and do not require an object to be invoked.

Example:

2.  Explain the concept of function arguments and parameters in Python.

=> **Parameters** are variables defined in the function's declaration (or definition). They act as placeholders for the values that the function will receive when called.

Example:

def greet(name):  # 'name' is a parameter

**Arguments** are the actual values passed to the function when it is called. These values replace the parameters in the function.

Example:

greet("Aishwarya")  # "Aishwarya" is an argument

4. What is the purpose of the `return` statement in a Python function?
=>
-It immediately terminates the execution of the function in which it is called. Any code after the return statement within the function is not executed.

-It allows the function to pass a value (or multiple values) back to the code that called it. This makes the function reusable and enables further processing of the returned value.

-By returning values, functions can perform specific tasks and provide results without directly interacting with the rest of the program, making the code cleaner and easier to maintain.

In contrast to print(), which displays output to the console, return provides a value that can be stored, manipulated, or used elsewhere in the program.

For example:
def add_numbers(a, b):
    return a + b  # Returns the sum of a and b
    

result = add_numbers(5, 3)
print(result)  # Output: 8

5. What are iterators in Python and how do they differ from iterables?
=>

**Iterables**
An iterable is any object that can be looped over (iterated through). Examples include lists, tuples, dictionaries, sets, and strings.
These objects implement the __iter__() method, which returns an iterator.
You can use a for loop or pass them to functions like iter() to retrieve their elements one by one.

**Iterators**
An iterator is an object that represents a stream of data. It is used to iterate over an iterable object.
Iterators implement two methods:
__iter__() – Returns the iterator object itself.
__next__() – Returns the next element in the sequence. When there are no more elements, it raises a StopIteration exception.
You can create an iterator by passing an iterable to the iter() function.

**Example**

Iterable

my_list = [1, 2, 3]  # A list is an iterable

Creating Iterator

my_iterator = iter(my_list)

print(next(my_iterator))  # Output: 1

print(next(my_iterator))  # Output: 2

print(next(my_iterator))  # Output: 3

 -iterables are like containers of data, while iterators are tools to sequentially access the data in those containers.

7. What are the advantages of using generators over regular functions?

=>

1. Memory Efficiency
Generators produce values one at a time using the yield statement, rather than creating and storing the entire dataset in memory. This is especially useful for handling large datasets or infinite sequences.
2. Lazy Evaluation
Generators compute values on demand, meaning they only generate the next value when requested. This reduces unnecessary computations and improves performance in scenarios where not all results are needed.
3. State Retention
Generators maintain their state between iterations. This allows them to resume execution right after the last yield statement, making them ideal for tasks like reading large files or streaming data.
4. Improved Readability
Generators simplify code for iterative processes by eliminating the need for managing intermediate states or manually implementing iterators.
5. Pipeline Creation
Generators can be chained together to create efficient data pipelines, processing data step-by-step without loading everything into memory.
6. Reduced Overhead
Unlike regular functions that return entire data structures, generators avoid the overhead of creating and returning large collections, leading to faster execution in many cases.

Example:


**Regular Function**

def square_numbers(nums):
    return [n * n for n in nums]

**Generator function**

def square_numbers_gen(nums):
    for n in nums:
        yield n * n

nums = range(1, 1000000)
gen = square_numbers_gen(nums)  # Memory-efficient
result = square_numbers(nums)  # Consumes more memory


Generators are a powerful tool for optimizing performance and memory usage, making them a preferred choice for many iterative and data-intensive tasks.

10.  What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?

=>
1. map()
Purpose: Applies a given function to each item in an iterable (e.g., list, tuple) and returns a new iterable with the results.
Use Case: Transforming or modifying data.

Example:


numbers = [1, 2, 3, 4]

squared = map(lambda x: x**2, numbers)

print(list(squared))  # Output: [1, 4, 9, 16]

2. filter()Purpose: Filters elements from an iterable based on a condition (a function that returns True or False).
Use Case: Selecting specific elements that meet a condition.

Example

numbers = [1, 2, 3, 4]

even_numbers = filter(lambda x: x % 2 == 0, numbers)

print(list(even_numbers))  # Output: [2, 4]


3. reduce()Purpose: Applies a function cumulatively to the items in an iterable, reducing it to a single value.
Use Case: Aggregating data (e.g., summing, multiplying).

from functools import reduce

numbers = [1, 2, 3, 4]

product = reduce(lambda x, y: x * y, numbers)

print(product)  # Output: 24

1.  Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in
the list.


```python
def sum_of_even_numbers(numbers):
    return sum(num for num in numbers if num % 2 == 0)

example_list = [1, 2, 3, 4, 5, 6]
result = sum_of_even_numbers(example_list)
print("Sum of even numbers:", result)
```

6. Write a generator function in Python that yields the powers of 2 up to a given exponent.

```python
def powers_of_two(max_exponent):
    """Generator that yields powers of 2 up to 2^max_exponent."""
    for exponent in range(max_exponent + 1):
        yield 2 ** exponent

for power in powers_of_two(5):
    print(power)
```

10. Create a Python program that uses `filter()` to remove all the vowels from a given string

```python
def remove_vowels(input_string):
    vowels = 'aeiouAEIOU'
    result = ''.join(filter(lambda char: char not in vowels, input_string))
    return result

input_string = "Hello, how are you?"
output_string = remove_vowels(input_string)
print("String without vowels:", output_string)
```
